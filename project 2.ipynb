```python
import pandas as pd
```

```python
data = pd.read_csv("james_bond_data_cleansed.csv")
data.head()
```

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = np.array(data["imdb"])
y = np.array(data["rotten_tomatoes"])
plt.scatter(x,y)
plt.xlabel("IMBD_rating")
plt.ylabel("Rotten Tomatoes")
plt.title("IMBD vs Rotten Tomatoes")
plt.savefig("IMBD vs Rotten Tomatoes.jpg")
plt.show()
# there seems to be a correlation beteen a higher rotten tomato score with a higher imbd score. It appears to be linear 
```

```python
x = np.array(data["imdb"])
y = np.array(data["bond_kills"])
plt.scatter(x,y)
plt.xlabel("IMBD_rating")
plt.ylabel("Bond_kills")
plt.title("IMBD vs Bond_kills")
plt.savefig("IMBD vs Bond_kills.jpg")
plt.show()

#There doesnt seem to be a coorelation beween more kills and a highr imbd rating 
```

```python
import matplotlib.pyplot as plt 
```

```python
x = data["release_date"]
y= (data["movie_budget"])
y1= (data["income_world"]) 

plt.plot(x, y, 'b-o', label= "Movie Budget")
plt.plot(x, y1, 'r-o', label= "World Revenue")
plt.xlabel("Year Release")
plt.legend()
plt.show()

# As time went on more money was put into the movies creating a large diffrence in income in the world over time.
```

```python
year = data["release_date"]
income = (data["income_world"])
budget = (data["movie_budget"])
ratio = (income/budget)

fig = plt.figure(figsize= (7,5))

plt.bar(year, ratio, color='blue', width= .5)
plt.xlabel(" Release Year")
plt.ylabel("Income world-Budget ratio")
plt.title("Income world over time")

#It seems that over time the income world and budget ratio went down as time went on.
```
